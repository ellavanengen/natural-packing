```python
def getPrecinctVotes(rootDir, year, countyName, election):
    #rootDir should be CountyVoteData
    #returns a dictionary from precicntGEOID --> {"DEM": demVotes, 
    #                                           "REP": repVotes, "TOTAL": totalVotes}
    if rootDir == "":
        rootDir = "./Downloads/CountyVoteData/"
    import os
    import pandas as pd

    if 'NEWHANOVER' in countyName:
        countyName = "NEW HANOVER"
    countyDir = os.path.join(rootDir, year, "PrecinctVotes"+election, countyName.upper()+".txt")
    if not os.path.exists(countyDir):
        print("ERROR: No data found for", countyName.upper(), 'county for the election', election, 'in the year', year)
    preDf = pd.read_csv(countyDir, sep = '\t')

    pToVotes = {}
    for p in preDf.PRECINCT.unique():
        soloPdf = preDf[preDf.PRECINCT == p]
        dem = int(sum(soloPdf.DEM))
        rep = int(sum(soloPdf.REP))
        total = int(sum(soloPdf.TOTAL))
        pToVotes[p] = {'DEM': dem, "REP": rep, "TOTAL": total}
    return pToVotes
```
